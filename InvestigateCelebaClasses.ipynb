```python
dataframe = pd.read_csv('list_attr_celeba.txt', delim_whitespace=True)
dataframe
```

```python
column_names = [
    'img_name',
    'Attractive_only',
    'Mouth_Slightly_Open_only',
    'Smiling_only',
    'Attractive_Mouth_Slightly_Open',
    'Attractive_Smiling',
    'Mouth_Slightly_Open_Smiling',
    'Attractive_Mouth_Slightly_Open_Smiling',
    'None'
]
```

```python
dataframe['Attractive_only'] = np.where(((dataframe.Attractive == 1) & 
                                          (dataframe.Mouth_Slightly_Open == -1) &
                                          (dataframe.Smiling == -1)), 1, 0)
dataframe['Mouth_Slightly_Open_only'] = np.where(((dataframe.Attractive == -1) & 
                                                  (dataframe.Mouth_Slightly_Open == 1) &
                                                  (dataframe.Smiling == -1)), 1, 0)
dataframe['Smiling_only'] = np.where(((dataframe.Attractive == -1) & 
                                      (dataframe.Mouth_Slightly_Open == -1) &
                                      (dataframe.Smiling == 1)), 1, 0)
dataframe['Attractive_Mouth_Slightly_Open'] = np.where(((dataframe.Attractive == 1) & 
                                                        (dataframe.Mouth_Slightly_Open == 1) &
                                                        (dataframe.Smiling == -1)), 1, 0)
dataframe['Attractive_Smiling'] = np.where(((dataframe.Attractive == 1) & 
                                            (dataframe.Mouth_Slightly_Open == -1) &
                                            (dataframe.Smiling == 1)), 1, 0)
dataframe['Mouth_Slightly_Open_Smiling'] = np.where(((dataframe.Attractive == -1) & 
                                                     (dataframe.Mouth_Slightly_Open == 1) &
                                                     (dataframe.Smiling == 1)), 1, 0)
dataframe['Attractive_Mouth_Slightly_Open_Smiling'] = np.where(((dataframe.Attractive == 1) & 
                                                                (dataframe.Mouth_Slightly_Open == 1) &
                                                                (dataframe.Smiling == 1)), 1, 0)
dataframe['None'] = np.where(((dataframe.Attractive == -1) & 
                               (dataframe.Mouth_Slightly_Open == -1) &
                               (dataframe.Smiling == -1)), 1, 0)
```

```python
dataframe_less = dataframe[column_names]
dataframe_less
```

```python

from skimage import io, transform
class FaceDataset(Dataset):
    """Face Landmarks dataset."""

    def __init__(self, csv_file, root_dir, transform=None):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        self.df = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        print(self.df['img_name'][idx])
        img_name = os.path.join(self.root_dir,
                                self.df['img_name'][idx])
        image = io.imread(img_name)
        image = np.moveaxis(image, -1, 0)
        print(image.shape)
        image = torch.from_numpy(image)
        features = self.df.iloc[idx, 1:]
        features = np.array([features])

        if self.transform:
            image = self.transform(image)
            
        sample = {'image': image, 'features': features}

        return sample
    
dataset = FaceDataset(csv_file='reduced_attr.csv',
                                    root_dir='./data')
```

```python
import torchvision.transforms as transforms
transform=transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize(64),
    transforms.CenterCrop(64),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),])

dataset = FaceDataset('data/celeba_smaller/reduced_attr.csv',
                      'data/celeba_smaller/img_align_celeba/',
                      transform=transform)
```
